```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from datetime import datetime

clean_data = Path(r"..\Resources\cleaned_netflix_df.csv")

netflix_data_df = pd.read_csv(clean_data)

netflix_data_df.head()
```

```python
#netflix_data_df['date_added'] = pd.to_datetime(netflix_data_df['date_added'])
netflix_data_df['date_added'] = pd.to_datetime(netflix_data_df['date_added'], format="%B %d, %Y", errors='coerce')
```

```python
netflix_data_df['year_added'] = netflix_data_df['date_added'].dt.year
netflix_data_df['month_added'] = netflix_data_df['date_added'].dt.month_name()
```

```python
release_counts = netflix_data_df.groupby(['year_added', 'month_added']).size().reset_index(name='count')

release_counts
```

```python
plt.figure(figsize=(12, 6))
plt.plot(release_counts['year_added'], release_counts['count'], marker='o')
plt.title('Netflix Content Releasement Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(release_counts['year_added'], release_counts['count'])
plt.title('Netflix Content Releasement Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(release_counts['month_added'], release_counts['count'], marker='o')
plt.title('Netflix Content Releasement Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Releases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
