```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load
netflix_titles_to_load = Path("Resources/netflix_titles.csv")

# Read file and store into Pandas DataFrame
orig_netflix_titles_df = pd.read_csv(netflix_titles_to_load)
orig_netflix_titles_df
```

```python
orig_netflix_titles_df = orig_netflix_titles_df.drop_duplicates()
orig_netflix_titles_df
```

```python
orig_netflix_titles_df = orig_netflix_titles_df.drop(columns=["show_id", "title", "director", "cast", "duration", "description"])
orig_netflix_titles_df
```

```python
orig_netflix_titles_df = orig_netflix_titles_df.dropna(subset=["country"])
orig_netflix_titles_df
```

```python
orig_netflix_titles_df = orig_netflix_titles_df.dropna()
orig_netflix_titles_df
```

```python
orig_netflix_titles_df.to_csv("Resources/cleaned_netflix_df.csv")
```
